```python
#This file contains the interaction between msm and msw variables

import pandas as pd
df_msm = pd.read_csv('msm_tt_rel_abundance_r.csv')
df_msw = pd.read_csv('msw_tt_rel_abundance_r.csv')


df_msm.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fig = plt.figure()
fig.subplots_adjust(hspace=2.0,wspace=2.0)



plt.subplot(521)
sns.distplot(df_msm['d5__corynebacterium1'])
sns.distplot(df_msw['d5__corynebacterium1'])



plt.subplot(522)
sns.distplot(df_msm['d5__lactobacillus'])
sns.distplot(df_msw['d5__lactobacillus'])

plt.subplot(523)
sns.distplot(df_msm['d5__gardnerella'])
sns.distplot(df_msw['d5__gardnerella'])

plt.subplot(524)
sns.distplot(df_msm['d5__sneathia'])
sns.distplot(df_msw['d5__sneathia'])

plt.subplot(525)
sns.distplot(df_msm['d5__finegoldia'])
sns.distplot(df_msw['d5__finegoldia'])

plt.subplot(526)
sns.distplot(df_msm['d5__anaerococcus'])
sns.distplot(df_msw['d5__anaerococcus'])

plt.subplot(527)
sns.distplot(df_msm['d5__peptoniphilus'])
sns.distplot(df_msw['d5__peptoniphilus'])

plt.subplot(528)
sns.distplot(df_msm['d5__staphylococcus'])
sns.distplot(df_msw['d5__staphylococcus'])

plt.subplot(529)
sns.distplot(df_msm['d5__streptococcus'])
sns.distplot(df_msw['d5__streptococcus'])

plt.subplot(5,2,10)
sns.distplot(df_msm['d5__veillonella'])
sns.distplot(df_msw['d5__veillonella'])


plt.show()
```
