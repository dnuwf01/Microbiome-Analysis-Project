```python
'''
Read the msm and msw data set and print the modified datasets containing the relative abundance
'''

import pandas as pd


msm = pd.read_csv('MSM43.csv')
msw = pd.read_csv('msw43.csv')

feature_list = ['d5__lactobacillus','d5__corynebacterium1','d5__gardnerella','d5__sneathia','d5__finegoldia','d5__anaerococcus','d5__peptoniphilus','d5__staphylococcus','d5__streptococcus',
            'd5__veillonella','Other']

msm = msm[feature_list]
msw = msw[feature_list]
```

```python
## calculate the relative abundance of each taxa and map to a csv file.
## msm

for i in range(43):
    total1 = msm.loc[i,:].sum()
    for j in range(11):
            msm.iloc[i,j] = msm.iloc[i,j]/total1
            

msm.to_csv('msm_tt_rel_abundance.csv')
    
```

```python
# msw

## calculate the relative abundance of each taxa and map to a csv file.
## msm

for i in range(43):
    total2 = msw.loc[i,:].sum()
    for j in range(11):
            msw.iloc[i,j] = msw.iloc[i,j]/total2
#msw
msw.to_csv('msw_tt_rel_abundance.csv')
   
```
