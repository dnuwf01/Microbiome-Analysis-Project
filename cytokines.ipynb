```python
#creating a dummy variable 


## a function that iterates over all the columns and creates a z_score 

def create_zscore():
    for col in df.columns:
        col_zscore = col + '_zscore'
        df[col_zscore] = (df[col] - df[col].mean())/(df[col].std(ddof=0))
        

create_zscore()    
df

```

```python
## convert the corresponding z-scores to probabilistic value/percentile values.

from scipy import stats

def create_perc_score():
    for j in df.columns:
        if not j.endswith('_zscore'):
            col_per_score = j + '_per_score'
            array = df[j]
            per_score = []
            for index in df[j]:
                per_score.append(stats.percentileofscore(array,index))
                df[col_per_score] = pd.Series(per_score)


                
                
                
create_perc_score()


df
# ar = df['tnfa']
# for j in df['tnfa']:
#     print("{:.3f}".format(stats.percentileofscore(ar,j)))
```

```python
### this block creates a composite score for all the outcomes
import numpy as np

zero_d = np.zeros(43)

#def comp_score():
lis = [j for j in df.columns if j.endswith('_per_score')]
df['comp_score'] = pd.Series(zero_d)
for colum in lis:
    df['comp_score'] = df['comp_score'] + df[colum]

    
df
            
                
```
