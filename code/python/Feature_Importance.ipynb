```python
'''
Read the msm top features and the msw top features in separate dataframes
Include an 'Other' column
Include a 'Target' column
Merge the two dataframes by columns
Classify Feature Importance using Random Forest.

'''
```

```python
#This file contains the interaction between msm and msw variables

import pandas as pd
df_msm = pd.read_csv('msm_tt_rel_abundance_r.csv')
df_msw = pd.read_csv('msw_tt_rel_abundance_r.csv')
df_msm.head()
```

```python
## modify the df_msm dataframe

feature_list = ['d5__lactobacillus','d5__corynebacterium1','d5__gardnerella','d5__sneathia','d5__finegoldia','d5__anaerococcus','d5__peptoniphilus','d5__staphylococcus','d5__streptococcus',
            'd5__veillonella','Other','Target']  

df_msm['Target']=1
df_msm_mod = df_msm[feature_list]
df_msm_mod.head()
```

```python
## modify the df_msw dataframe

feature_list = ['d5__lactobacillus','d5__corynebacterium1','d5__gardnerella','d5__sneathia','d5__finegoldia','d5__anaerococcus','d5__peptoniphilus','d5__staphylococcus','d5__streptococcus',
            'd5__veillonella','Other','Target']  

df_msw['Target']=0
df_msw_mod = df_msw[feature_list]
df_msw_mod.head()
```

```python
## merge the two dataframes by columns

result = df_msm_mod.append(df_msw_mod)
```

```python
result = result.sample(frac=1)
y = result['Target']
X = result.drop(['Target'],axis=1)


result.to_csv('combined_msm_msw.csv')
X.head()
```

```python
##### slight change to the code
## we insert the combined log transformed csv here

ddf = pd.read_csv('bacteria_lt.csv')
ddf = ddf.sample(frac=1)
ddf.columns
```

```python
### create the X and y set

feature_list = ['d5__lactobacillus','d5__corynebacterium1','d5__gardnerella','d5__sneathia','d5__finegoldia','d5__anaerococcus','d5__peptoniphilus','d5__staphylococcus','d5__streptococcus',
            'd5__veillonella','Other'] 

X = ddf[feature_list]
y = ddf['status']
```

```python
## split the data set

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)

from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
#scores = cross_val_score(clf,X,y,cv=5)

clf.fit(x_train,y_train)
#y_pred = clf.predict(x_test)

print(clf.score(x_test,y_test))
# print(confusion_matrix(y_test,y_pred))


#print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
import numpy as np

lreg = LogisticRegression(solver='lbfgs').fit(x_train,y_train)
#scores = cross_val_score(lreg,X,y,cv=5)
# scores
print(lreg.score(x_test,y_test))
# y_pred = lreg.predict(x_test)
# print(confusion_matrix(y_test,y_pred))

#print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

abs(lreg.coef_*100)



#print(np.std(X, 0)*lreg.coef_[0][0])

#print(np.std(X,0)*lreg.coef_[0])

```
